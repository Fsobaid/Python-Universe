***We have a text file of time and x,y and z coordinates. We want to import them and save each of them into a list. Next, we wanna export them back into another text file as they were in the original file.<br>
You may ask why we do that, for practice purposes.***

```python

%matplotlib inline
import numpy as np
from scipy.special import factorial, gamma
import scipy
import matplotlib.pyplot as plt
from random import randint
import math as ma
# read from the file
f = open('mc_xyz11.txt', 'r')

# extract the contents from the txt file
content = f.readlines()

# open list variables to store the values in them
t = []
RX = []
RY = []
RZ = []

# loop to go throught the values in the file
for i in content:
    a,b,c,d = i.split()  #a,b,c,d are the columns, we nammed them you could choose other letters
    t.append(int(a))     # append means add the values to the list
    RX.append(float(b))   # float because the values with dicimals
    RY.append(float(c))
    RZ.append(float(d))
```

```python
plt.plot(t,RX,'r')
```

```python
f=open('mc35.txt', 'w+')

for l, x, j, k in zip(t,RX, RY, RZ):
    f.write(f'{l} {x} {j} {k}\n')
```
